```python
def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.01)
    quartile3 = dataframe[variable].quantile(0.99)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
```

```python
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

```python
def retail_data_prep(dataframe):
    dataframe.dropna(inplace=True)
    dataframe = dataframe[dataframe["Quantity"] > 0]
    dataframe = dataframe[dataframe["price"] > 0]
    replace_with_thresholds(dataframe, "Quantity")
    replace_with_thresholds(dataframe, "price")
    return dataframe
```

```python
#fonksiyonlaştırma
def create_invoice_product_df(dataframe, id=False):
    if id:
        return dataframe.groupby(['user_session', "product_id"])['Quantity'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
    else:
        return dataframe.groupby(['user_session', 'Description'])['Quantity'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
```

```python
fr_inv_pro_df = create_invoice_product_df(df)

fr_inv_pro_df = create_invoice_product_df(df, id=True)
```

## Association Rules

```python
frequent_itemsets = apriori(fr_inv_pro_df,
                            min_support=0.0001,
                            use_colnames=True)
```

```python
frequent_itemsets.sort_values("support", ascending=False)
```

```python
rules = association_rules(frequent_itemsets,
                          metric="support",
                          min_threshold=0.1)
```

```python
rules
```

```python
rules[(rules["support"]>0.05) & (rules["confidence"]>0.1) & (rules["lift"]>5)]
```

```python
rules[(rules["support"]>0.05) & (rules["confidence"]>0.1) & (rules["lift"]>5)]. \
sort_values("confidence", ascending=False)
```

## Ürün Önerisinde Bulunma

```python
product_id = 1002532
check_id(df, product_id)
```

```python
sorted_rules = rules.sort_values("lift", ascending=False)
```

```python
recommendation_list = []
```

```python
for i, product in enumerate(sorted_rules["antecedents"]):
    for j in list(product):
        if j == product_id:
            recommendation_list.append(list(sorted_rules.iloc[i]["consequents"])[0])
```
