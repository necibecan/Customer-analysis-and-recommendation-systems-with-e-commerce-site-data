```python
# skor sahibi olan kullanıcıları sınırlandırma
def create_users_df():

    comment_counts1 = pd.DataFrame(df["user_id"].value_counts())
    rare_product1 = comment_counts1[comment_counts1["user_id"] <= 100].index 
    common_product1 = df[~df["user_id"].isin(rare_product1)]
    user_df = common_product1.pivot_table(index=["user_id"], columns=["product_id"], values="user_score")
    
    return user_df
```

```python
user_df = create_users_df()
user_df.head()
```
